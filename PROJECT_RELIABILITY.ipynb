# Calculating Cronbach's alpha

```python
#CONVERSION OF FUZZY VARIABLES INTO CRISP DATA

def conversion_to_crisp(fuzzy_set):
     
    Crisp_value= np.mean(fuzzy_set, dtype=np.float64)
    return Crisp_value
Crisp_values=np.ones(7)
for i in range(7):
    Crisp_values[i]=conversion_to_crisp(np.array(fmf_dataset.iloc[1:,i]))
 
print(Crisp_values)

```

```python
array=np.arange(1,8,1)
print(array)

Crisp_values_reliability_dataset=Reliability.replace(array,Crisp_values)   
Crisp_values_reliability_dataset.sample(5)
```

```python
Crisp_dataset=Crisp_values_reliability_dataset.apply(np.sqrt) 
Crisp_dataset.sample(5)
```

```python
Question1=Question.replace(array,Crisp_values)
Crisp=Crisp_dataset.div(3)
Crisp.sample(5)
```

```python
matrix=pd.DataFrame(Crisp.values*Question1.values, columns=Crisp.columns, index=Crisp.index)
matrix.sample(5)
```

```python
TMC=matrix.iloc[:,:4].mean(axis=1)
LR=matrix.iloc[:,4:7].mean(axis=1)
FL=matrix.iloc[:,7:11].mean(axis=1)
RC=matrix.iloc[:,11:14].mean(axis=1)
AW=matrix.iloc[:,14:19].mean(axis=1)
PR=matrix.iloc[:,19:23].mean(axis=1)
RD=matrix.iloc[:,23:25].mean(axis=1)

H=matrix.iloc[:,25:30].mean(axis=1) 
S=matrix.iloc[:,30:33].mean(axis=1)
E=matrix.iloc[:,33:38].mean(axis=1)
ER=matrix.iloc[:,38:39].mean(axis=1)

data=[TMC,LR,FL,RC,AW,PR,RD,H,S,E,ER]
IMF=pd.DataFrame(data)
IMF.T.sample(5)
```
